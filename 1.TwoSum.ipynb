```python
import numpy as np
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        # Using the numpy package to do element-wise subtraction between lists
        rev = np.subtract([target] * len(nums), nums)
        # Init an empty list to store the indices
        loc = []
        # Iterative loop
        for i in nums:
            # Checking the number of occurances after subtraction in the list
            # Cannot let the element that is the half of our target pass through the if operator
            if np.count_nonzero(rev == i) == 1 and target - i != i:
                # use the np.where() function to check where the element is in our array after subtratcion
                # the result will be a tuple, retreving the first-element inside the tuple
                ind = np.where(rev == i)[0]
                # store it
                loc.append(int(ind))
            # If there are two numbers are exactly the half of our target
            elif np.count_nonzero(rev == i) == 2:
                # store it
                ind = np.where(rev == i)[0].tolist()
                loc.append(ind)
        # reverse the indices 
        loc.reverse()
        if isinstance(loc[0], list):
            return(loc[0])
        else:
            return(loc)
```

```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        # Init a dictionary
        d = {}
        # i for index, j for value
        for i, j in enumerate(nums):
            r = target - j # get the difference between the target and the element
            # if the residual is inside the dictionary, retrieve the index stored in the dictionary,
            # It's because the residual + j = target, it will always give the opposite value of it.
            if r in d: return [d[r], i]
            # store the value and the index inside the dict
            d[j] = i
```
